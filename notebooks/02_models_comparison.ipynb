```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../runs/models_comparison.csv")
df = df.sort_values(by="f1", ascending=False)
df
```

```python
df.plot(x="model", y="f1", kind="bar", figsize=(10, 5), legend=False)
plt.title("Сравнение моделей по F1-score")
plt.xlabel("Модель")
plt.ylabel("F1-score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.plot(x="model", y="fps", kind="bar", figsize=(10, 5), legend=False)
plt.title("Сравнение моделей по скорости инференса")
plt.xlabel("Модель")
plt.ylabel("FPS")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
